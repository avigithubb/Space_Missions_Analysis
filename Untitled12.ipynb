```python
# launch_count = yearsly_df.groupby(["Years", "Organisation"]).size().reset_index(name="Launch Count")

Luanch_counts = df.groupby(['Years', 'Organisation']).size().reset_index(name='Launch Count')
highest_launches = Luanch_counts.groupby("Years")["Launch Count"].max()
result = pd.merge(Luanch_counts, highest_launches, on=['Years', 'Launch Count'])

```

```python
pivote_df = result.pivot(index="Years", columns="Organisation", values="Launch Count")
pivote_df
```

```python
import matplotlib.pyplot as plt
pivote_df.plot(kind="bar", stacked=True, figsize=(20, 20))

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Years", fontsize=14)
plt.ylabel("Space Missions", fontsize=14)
plt.legend(title="Organisation", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```

```python
# By the looks of the bar chart we can conclude that most of the space missions has done by RVSN USSR so far.
# Strategic Missile Forces of the Russian Federation is a separate-troops branch of the Russian Armed Forces that controls Russia's land-based intercontinental ballistic missiles (ICBMs).
# It was formerly part of the Soviet Armed Forces from 1959 to 1991. The USSR was formally dissolved on December 26, 1991, and the Soviet hammer and sickle flag was replaced by the Russian tricolor over the Kremlin.
```

```python
launch_month = df.groupby(["Organisation", "Months"]).size().reset_index(name="Launch Month")
```

```python
another_pivote_df = launch_month.pivot(index="Months", columns="Organisation", values="Launch Month")
```

```python
another_pivote_df.plot(kind="bar", stacked=True, figsize=(20, 20))

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Years", fontsize=14)
plt.ylabel("Launch Months", fontsize=14)
plt.legend(title="Organisation", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```

```python
# The bar chart analysis between the launch months and the organizations reveals a significant trend.
# It appears that December has consistently been the most favorable month for space missions.
# This trend may be attributed to stable weather conditions and clear atmospheric conditions during this time of year.
# As a result, a higher concentration of space missions occurred in December, making it a preferred month for launching missions across various organizations.
```

```python
success_rate = []

for i in df.Mission_Status:
  if i == "Success":
    success_rate.append(1)
  elif i == "Failure":
    success_rate.append(0)
  else:
    success_rate.append("Nan")
```

```python
df.insert(6, "Success_Rate", success_rate)
```

```python
df
```

```python
Success_df = df[["Organisation", "Years", "Success_Rate"]]
```

```python
Success_df
```

```python
reshaped_df = df.groupby(["Years", "Success_Rate"]).size().reset_index(name="Final count")
reshaped_df
```

```python
pivoted_success = reshaped_df.pivot(index="Success_Rate", columns="Years", values="Final count")
```

```python
pivoted_success
rebased_success = pivoted_success.drop('Nan')
# rebased_success.loc[0]

row_0 = rebased_success.loc[0]

row_1 = rebased_success.loc[1]
```

```python
plt.figure(figsize=(20, 20))
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Years", fontsize=14)
plt.ylabel("Success_Rate", fontsize=14)
# rebased_success.index[0]
plt.ylim(0, 200)

plt.plot(row_0.index, row_0.values, marker="o")
plt.plot(row_1.index, row_1.values, marker="o")
plt.grid(True)



plt.show()

# for column in pivoted_success.columns:
  # plt.plot(pivoted_success.Success_Rate, pivoted_success[column])
  # print(column)
# plt.plot(pivoted_success.index, pivoted_success[1957])
```
